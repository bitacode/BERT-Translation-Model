**<h1>SPLIT DATASET AND RESTRUCTURE DICTIONARY</h1>**

```python
train_ratio = 0.7
test_ratio = 0.2
validation_ratio = 0.1
```

```python
train_data, temp_data = train_test_split(QED, test_size=1 - train_ratio, random_state=42)
test_data, validation_data = train_test_split(temp_data, test_size=test_ratio/(test_ratio + validation_ratio), random_state=42)
```

```python
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)
validation_data = validation_data.reset_index(drop=True)
```

```python
labeled_train_data = {'train': [{'id': row['text_id'], 'en': row['text_en']} for index, row in train_data.iterrows()]}
```

```python
labeled_test_data = {'test': [{'id': row['text_id'], 'en': row['text_en']} for index, row in test_data.iterrows()]}
```

```python
labeled_validation_data = {'validation': [{'id': row['text_id'], 'en': row['text_en']} for index, row in validation_data.iterrows()]}
```

```python
temp_dict = {**labeled_train_data, **labeled_test_data, **labeled_validation_data}
```

```python
temp_datasets = DatasetDict(temp_dict)
```

```python
raw_datasets = DatasetDict({})
for dataset_name, dataset in temp_datasets.items():
    translation_feature = {'translation': [{'id': ex['id'], 'en': ex['en']} for ex in dataset]}
    raw_datasets[dataset_name] = Dataset.from_dict(translation_feature)
```

**<h1>BUILDING MODEL</h1>**

```python
model_checkpoint = 'Helsinki-NLP/opus-mt-id-en'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
with tokenizer.as_target_tokenizer():
    print(tokenizer(['Ada perasaan ikatan.']))
```

```python
max_input_length = 128
max_target_length = 128

source_lang = 'id'
target_lang = 'en'

def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples['translation']]
    targets = [ex[target_lang] for ex in examples['translation']]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)

    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
preprocess_function(raw_datasets['train'][:2])
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

```python
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
batch_size = 16
learning_rate = 2e-5
weight_decay = 0.01
num_train_epochs = 10 # keep adjusting iteration number until you get desired result
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors='tf')
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors='tf', pad_to_multiple_of=128)
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets['test'],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets['validation'],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets['validation'],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=num_train_epochs)
```

```python
model.save_pretrained('translation_model/')
```
